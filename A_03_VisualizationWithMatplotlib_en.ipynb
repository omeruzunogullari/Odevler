In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
df.columns
```

```python
sp_df = df[df.location =="Spain"][["date", "total_cases", "total_deaths", "total_cases_per_million"]]
fr_df = df[df.location =="France"][["date", "total_cases", "total_deaths", "total_cases_per_million"]]
ge_df = df[df.location =="Germany"][["date", "total_cases", "total_deaths", "total_cases_per_million"]]
it_df = df[df.location =="Italy"][["date", "total_cases", "total_deaths", "total_cases_per_million"]]

title_style = {"family": "Arial", "color": "darkred", "size": 20}
axis_style = {"family": "Arial", "color": "darkred", "size": 20}

days = -125 ## Günleri neden -125'ten başlattığımızı anlamadım.

plt.figure(figsize=(20,8))
plt.suptitle = ("Coronavirus Total Cases and Total Deaths", )

plt.subplot(1,2,1)
plt.plot(sp_df.date[days:], sp_df.total_cases[days:], label="Spain", lw=3)
plt.plot(fr_df.date[days:], fr_df.total_cases[days:], label="France", lw=3)
plt.plot(ge_df.date[days:], ge_df.total_cases[days:], label="Germany",lw=3)
plt.plot(it_df.date[days:], it_df.total_cases[days:], label="Italy", lw=3)

plt.title("Total Cases", fontdict=axis_style)
plt.xlabel("Dates", fontdict=axis_style)
plt.ylabel("Count of Cases", fontdict=axis_style)
plt.xticks(rotation="45")
plt.legend()

plt.subplot(1,2,2)
plt.plot(sp_df.date[days:], sp_df.total_deaths[days:], label="Spain", lw=3)
plt.plot(fr_df.date[days:], sp_df.total_deaths[days:], label="France", lw=3)
plt.plot(ge_df.date[days:], ge_df.total_deaths[days:], label="Germany", lw=3)
plt.plot(it_df.date[days:], it_df.total_deaths[days:], label="Italy", lw=3)

plt.title("Total Deaths", fontdict=title_style)
plt.xlabel("Dates", fontdict=axis_style)
plt.ylabel("Count of Deaths", fontdict=axis_style)
plt.xticks(rotation="45")
plt.legend()

plt.show()
```
